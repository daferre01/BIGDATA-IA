```python
import ccxt
import time
from datetime import datetime
import os
exchange = ccxt.binance()

symbol = 'BTC/USDT' 

def obtener_datos():
    try:
        os.system('cls' if os.name == 'nt' else 'clear')
        ordenes = exchange.fetch_order_book(symbol)
        precio_actual = exchange.fetch_ticker(symbol)['last']
        stop_loss = None
        take_profit = None
        tendencia = None
        dia = datetime.now().strftime('%Y-%m-%d')
        hora = time.strftime('%H:%M:%S')
        print("% P&G:", precio_actual)
        print("Dia:", dia)
        print("Hora:", hora)
        
    except Exception as e:
        print("Error al obtener datos:", e)

while True:
    obtener_datos()
    time.sleep(1)  
```
