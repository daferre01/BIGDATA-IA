```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread('imagenes/j0.png')
src2 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
plt.imshow(src2, cmap='gray')
```

```python
kernel = np.ones((5,5), np.uint8)
erosion= cv.erode(img1,kernel, iterations=1)
fig,(ax1,ax2)= plt.subplots(1,2)
ax1.imshow(img1,cmap='gray')
ax2.imshow(erosion,cmap='gray')
```

```python
kernel = np.ones((5,5), np.uint8)
dilatacion= cv.dilate(img1,kernel, iterations=1)
fig,(ax1,ax2)= plt.subplots(1,2)
ax1.imshow(img1,cmap='gray')
ax2.imshow(dilatacion,cmap='gray')
```

```python
img1 = cv.imread("imagenes/j2.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
closing = cv.morphologyEx(img1, cv.MORPH_CLOSE, kernel)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(closing, cmap='gray')
ax2.set_title("Closing")
```

```python
img1 = cv.imread("imagenes/j1.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
opening = cv.morphologyEx(img1, cv.MORPH_OPEN, kernel)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(opening, cmap='gray')
ax2.set_title("Opening")
```

```python
img1 = cv.imread("imagenes/j0.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)

kernel = np.ones((5,5),np.uint8)
closing = cv.distanceTransform(img1, cv.DIST_L2,0)

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(closing, cmap='gray')
ax2.set_title("Gradient")
```

```python
img1 = cv.imread("imagenes/letras.png")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)


kernel = np.ones((5,5),np.uint8)
img2 = cv.morphologyEx(img1, cv.MORPH_CLOSE, kernel, iterations=4) # limpiamos letras
img2 = cv.morphologyEx(img2, cv.MORPH_OPEN, kernel, iterations=3) # limpiamos fondo

fig, (ax1, ax2) = plt.subplots(1, 2)

fig.suptitle('')

ax1.imshow(img1, cmap='gray')
ax1.set_title("Org")
ax2.imshow(img2, cmap='gray')
ax2.set_title("Opening")
```
