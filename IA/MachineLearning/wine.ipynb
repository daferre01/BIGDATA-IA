# matriz de correlacion y mapa de calor

```python

# Matriz de correlacin
corr_matrix= datos.corr()
corr_matrix
```

```python
sorted(datos.quality.unique())
```

```python
# mapa de color
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15, 8))
text_colors = ("black" , "white" )
im = ax.imshow(corr_matrix, cmap= "Greens" ) # mapa de calor
cbar = fig.colorbar(im, ax=ax, label= "Correlacion" ) # leyenda
cbar.outline.set_visible(False)

x = corr_matrix.columns
y = corr_matrix.index
# Mostrar las etiquetas. El color del texto cambia en funcin de su normalizacin
for i in range(len(y)):
  for j in range(len(x)):
    value = corr_matrix.iloc[i, j]
    text_color = text_colors[int(im.norm(value) > 0.5)] # color etiqueta
    ax.text(j, i, f"{value:.2f}" , color=text_color, va= "center" , ha= "center" )



# Formateo de los ejes
ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation=90)
ax.set_yticks(range(len(y)))
ax.set_yticklabels(y)
ax.invert_yaxis()
ax.spines["right"].set_visible(False) # ocultar borde derecho
ax.spines["top"].set_visible(False) # ocultar borde superior
fig.tight_layout()
```

```python
def split_train_test(data, test_ratio):
  # indices = [i for i in range(len(data))]; indices = np.shuffle(.....)
  indices = np.random.permutation(len(data))
  lg_test = int(len(data) * test_ratio)
  test_indices = indices[:lg_test]
  train_indices = indices[lg_test:]

  return data.iloc[train_indices], data.iloc[test_indices]

np.random.seed(42)

datos_train, datos_test = split_train_test(datos, 0.2)
len(datos_train), len(datos_test)
```

```python
y_train = datos_train["quality"].copy()
x_train = datos_train.drop("quality", axis=1)
x_train = datos_train[["alcohol", "citric acid", "volatile acidity"]]
```

```python
from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
print(f'--------------quality--------')
#regresion lineal

datos_prueba=x_train.iloc[:10]
print('Prediciones:', lin_reg.predict(datos_prueba))
y_reales=y_train.iloc[:10]
print('Reales:', list(y_reales))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

predicciones = lin_reg.predict(x_train)
mse = mean_squared_error(y_train, predicciones)
mse = np.sqrt(mse)
mae = mean_absolute_error(y_train, predicciones)
score = r2_score(y_train, predicciones)
print(f"mae: {mae}   rmse: {mse} r2_score: {score}")
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(x_train, y_train)
```

```python
y_train2 = datos_train["quality"].copy()
x_train2 = datos_train.drop("quality", axis=1)
x_train2 = datos_train[["alcohol", "citric acid", "volatile acidity"]]
```

```python
predicciones = tree_reg.predict(x_train2)
mse = mean_squared_error(y_train2, predicciones)
mse = np.sqrt(mse)
mae = mean_absolute_error(y_train2, predicciones)
score = r2_score(y_train2, predicciones)
print(f"mae: {mae}   rmse: {mse} r2_score: {score}")
```

```python
# cross-validation arbol decision
from sklearn.model_selection import cross_val_score

lin_score = cross_val_score(tree_reg, x_train, y_train,
                            scoring = "neg_mean_squared_error", cv=10)
root_lin_score = np.sqrt(-lin_score)
print("Scores: ", root_lin_score)
print("Media: ", root_lin_score.mean())
print("Desviacin Std", root_lin_score.std())
```

```python
from sklearn.ensemble import RandomForestRegressor
y_test = datos_train["quality"].copy()
x_test = datos_train.drop("quality", axis=1)
x_test = datos_train[["alcohol", "citric acid", "volatile acidity"]]
rf_reg = RandomForestRegressor(n_estimators=100)

rf_reg.fit(x_test, y_test)

rf_score = cross_val_score(rf_reg, x_test, y_test,
                            scoring = "neg_mean_squared_error", cv=10)
root_lin_score = np.sqrt(-lin_score)
print("Scores: ", root_lin_score)
print("Media: ", root_lin_score.mean())
print("Desviacin Std", root_lin_score.std())
predicciones = rf_reg.predict(x_test)
mse = mean_squared_error(y_test, predicciones)
mse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predicciones)
score = r2_score(y_test, predicciones)
print(f"mae: {mae}   rmse: {mse} r2_score: {score}")
```

```python
import xgboost as xgb

xgb_reg = xgb.XGBRegressor(objetive="reg:squarederror")

xgb_reg.fit(x_train, y_train)

predicciones = xgb_reg.predict(x_test)
mse = mean_squared_error(y_test, predicciones)
mse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predicciones)
score = r2_score(y_test, predicciones)
print(f"mae: {mae}   rmse: {mse} r2_score: {score}")

rf_score = cross_val_score(xgb_reg, x_train, y_train,
                            scoring = "neg_mean_squared_error", cv=10)
root_lin_score = np.sqrt(-rf_score)
print("XGB cross")
print("Scores: ", root_lin_score)
print("Media: ", root_lin_score.mean())
print("Desviacin Std", root_lin_score.std())
```
